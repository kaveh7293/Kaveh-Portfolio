```python
data=pd.read_csv('WorldCup.csv')
data.head()
```

```python
X_all=[]
y_all=[]
j=0
# for loop to collect the winners in all the games
for i in range(188):
    
    xa=list(data.iloc[i,[1,2,3,4,6,7,8,9]].values)
    xa[0]=re.split("\'", xa[0])[0]
    xa[3]=re.split("\'", xa[4])[0]

    X_all.append(xa)
    y_all.append(data.iloc[i,11])
    
for i in range(188):
    xa=list(data.iloc[i,[6,7,8,9,1,2,3,4]].values)
    xa[0]=re.split("\'", xa[0])[0]
    xa[3]=re.split("\'", xa[4])[0]
    X_all.append(xa)
    y_all.append(2-data.iloc[i,11])
    
y_all=np.array(y_all)
y_all=y_all.reshape((-1,1))
```

```python
data_frame=pd.DataFrame(data=X_all,columns=['Continent1','rank1','Num_Champs1'
                                            ,'IsHost1','Continent2','rank2','Num_Champs2','IsHost2'])
data_frame2=pd.get_dummies(data_frame)
Xtrain,Xtest,Ytrain,Ytest=train_test_split(data_frame2.values,y_all,test_size=0.2,random_state=0)
SC=StandardScaler()
Xtrain_Scaled=SC.fit_transform(Xtrain)
data_frame2.columns
```

```python
shape_X=Xtrain_Scaled.shape
corr_vector=[]
for i in range(shape_X[1]):
    r1=np.corrcoef(Xtrain_Scaled[:,i],Ytrain.reshape((300,)))
    corr_vector.append(r1[0,1])
```

```python
plt.figure(figsize=(20,10))
x_label=list(range(1,len(corr_vector)+1))
x_thicks=list(range(1,len(corr_vector)+1,1
            ))
plt.bar(x_label,np.abs(corr_vector))
plt.xticks(x_thicks,rotation=90,family='Times New Roman',fontweight='bold',fontsize=18)
plt.yticks([0,0.05,0.1,0.15,0.2],family='Times New Roman',fontweight='bold',fontsize=28)
plt.xlabel('Features',family='Times New Roman',fontweight='bold',fontsize=25)
plt.ylabel('Correlation Coefficient',family='Times New Roman',fontweight='bold',fontsize=25)

plt.axhline(y=0.1,ls='--',c='r')
```
